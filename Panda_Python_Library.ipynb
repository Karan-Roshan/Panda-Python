# **Pandas - Cleaning Empty Cells**


```python
df = pd.read_csv('workout_data.csv')
```

```python
df
```

## **Empty Cells**

Empty cells can potentially give you a wrong result when you analyze data.



**.isnull()** : It is used to detect missing (null/NaN) values in a DataFrame or Series.

```python
df.isnull().sum()
```

## **Remove Rows**

One way to deal with empty cells is to **remove rows that contain empty cells**.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.



**dropna()**: Return a new Data Frame with no empty cells:

```python
new_df = df.dropna()

print(new_df.to_string())
```

**Note:** By default, the dropna() method returns a new DataFrame, and will not change the original.

```python
# drop any row that had any null value

df.dropna(how = "any")
```

```python
# if all the values in a row are null then we drop that row

df.dropna(how = "all")
```

If you want to change the original DataFrame, use the **inplace = Tru**e argument:

```python
df.dropna(inplace = True)

print(df.to_string())
```

**Note:** Now, the **dropna(inplace = True) will NOT return a new DataFrame**, but it will remove all rows containing NULL values from the original DataFrame.

## **Fill Null Value**

Filling Null Values means replacing missing values (NaN) with a specified value using the fillna() method.

```python
df = pd.read_csv('workout_data.csv')
```

```python
df.fillna(0)
```

```python
df["Calories"].fillna(df["Calories"].mean())
```

```python
df["Pulse"].fillna(df["Pulse"].median())
```

### **Data imputation methods or missing value handling techniques**

These methods are used to fill missing (NaN) values using nearby values.

1. Forward Fill
2. Backward Fill

#### **Forward Fill (ffill)**

Forward Fill replaces a missing value with the previous non-null value.

```python
df["Maxpulse"].fillna(method = "ffill")
```

#### **Backward Fill (ffill)**

Backward Fill replaces a missing value with the next non-null value.

```python
df["Maxpulse"].fillna(method = "bfill")
```

## **Replace Empty Values**

Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value.

```python
df = pd.read_csv('workout_data.csv')
```

```python
print(df.shape)
```

```python
df.fillna(150)
```

```python
df.fillna(130)
```

## **Replace Only For Specified Columns**

The example above replaces all empty cells in the whole Data Frame.

To only replace empty values for one column, specify the column name for the DataFrame:



```python
df.fillna({"Calories": 130}, inplace = True)

df
```

```python
df["Name"] = df["Name"].replace("Rahul", "Nyasha")

df
```

# **Removing Duplicates**

## **Duplicates**

Find the duplicate records in the DataFrame.

```python
df_dup = df[df.duplicated()]

df_dup
```

```python
# By default, df.duplicated() uses keep='first',
# which keeps the first occurrence of a duplicate row and marks subsequent occurrences as duplicates (True).

df_dup = df[df.duplicated(keep = "first")]

df_dup
```

```python
df_dup = df[df.duplicated(keep = "last")]

df_dup
```

## **Removing the Duplicates**

This statement removes duplicate rows from the DataFrame and stores the result back into df.

To remove duplicates, use the drop_duplicates() method.

**Remember:** The (inplace = True) will make sure that the method does NOT return a new DataFrame, but it will remove all duplicates from the original DataFrame.

# **Joins & Merges**

Merging and Joining are used to combine two or more DataFrames based on a common column or index.

```python
employees = {
    "Emp_ID": [101, 102, 103, 104, 105],
    "Name": ["Karan", "Rahul", "Aman", "Priya", "Sneha"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing"]
}

df1 = pd.DataFrame(employees)

df1
```

```python
salaries = {
    "Emp_ID": [101, 102, 104, 106],
    "Salary": [50000, 45000, 55000, 60000]
}

df2 = pd.DataFrame(salaries)

df2
```

## **Merge**

merge() combines DataFrames using one or more common columns (similar to SQL JOIN).

### **concat()**

concat() is used to combine (concatenate) two or more DataFrames or Series either:

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

### **merge()**

merge() combines DataFrames using one or more common columns

```python
pd.merge(df1, df2, on = "Emp_ID")
```

# **Pandas - Plotting**

## **Plotting**

Pandas uses the plot() method to create diagrams.

We can use Pyplot, a submodule of the Matplotlib library to visualize the diagram on the screen.

## **Scatter Plot**

Specify that you want a scatter plot with the kind argument:

**kind = 'scatter'**

A scatter plot needs an **x-axis** and a **y-axis.**

```python
# The fact that higher duration means more calories burned.

df.plot(kind = "scatter", x = "Duration", y="Calories")
```

```python
# A scatterplot where there are no relationship between the columns:

df.plot(kind = "scatter", x = "Duration", y="Maxpulse")
```

## **Histogram**

Use the kind argument to specify that you want a histogram:

**kind = 'hist'**

A histogram needs **only one column**.

A histogram shows us the frequency of each interval
